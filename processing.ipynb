```python
# استخراج محتوى العمود "data" من الصف رقم 3 (مثلاً index = 3)
data_list = df.loc[2, 'data']

# تحويله إلى DataFrame وأخذ فقط عمود "text"
df_text = pd.DataFrame(data_list)[['text']]

# إعادة تسمية العمود إلى "data"
df_text = df_text.rename(columns={'text': 'data'})

# عرض النتيجة
df_text

```

```python
# نفترض إن df_text فيه عمود واحد اسمه 'data' والنصوص جوا
# ناخذ النصوص كلها ونحولها إلى كلمات
words = " ".join(df_text['data']).split()

# نعمل DataFrame جديد فيه index تلقائي لكل كلمة
df_words = pd.DataFrame(words, columns=['word'])

# عرض النتيجة
df_words
```

```python
# دالتنا للفحص والتصنيف
def classify_word(word):
    tanwin = ["ً", "ٌ", "ٍ"]
    if isinstance(word, str):
        # إذا آخر حرف تنوين
        if len(word) >= 1 and word[-1] in tanwin:
            return "اسم"
        # إذا الحرف قبل الأخير تنوين وبعده ألف
        if len(word) >= 2 and word[-2] in tanwin and word[-1] == "ا":
            return "اسم"
    return None

# إنشاء عمود tag
df_words["tag"] = df_words["word"].apply(classify_word)

# إنشاء عمود المحذوف (الكلمات اللي صنفت كـ "اسم")
df_words["المحذوف"] = df_words.apply(
    lambda row: row["word"] if row["tag"] == "اسم" else None, axis=1
)

# حذف الكلمات المصنفة من العمود word (إبقائها None)
df_words.loc[df_words["tag"] == "اسم", "word"] = None

# لو بدك العمود "data" بدل "word"
df_words = df_words.rename(columns={"word": "data"})
```

```python
import pandas as pd
import re

# --- 1) اقرأ الملف df1 ---
df1 = pd.read_excel(r"C:\Users\Eqra\Desktop\Quran_Project\Data\Data\arabic_jamida_names.xlsx")

# --- 2) جهّز اسم العمود الصحيح (index = 2) ---
col_name = df1.columns[2]  # العمود الثالث
print("Using df1 column:", col_name)

# --- 3) دالة للتطبيع وإزالة التشكيل ---
def normalize_ar(word):
    """إزالة التشكيل والتطويل + تطبيع بعض الحروف"""
    if not isinstance(word, str):
        return ""
    s = word.strip()
    s = re.sub(r'[\u064B-\u0652\u0670\u0653\u0654\u0655\u0640]', '', s)  # إزالة التشكيل والتطويل
    s = s.replace('أ', 'ا').replace('إ', 'ا').replace('آ', 'ا')
    s = s.replace('ى', 'ي').replace('ؤ', 'و').replace('ئ', 'ي').replace('ة', 'ه')
    return s

# --- 4) جهّز مجموعة الأسماء من df1 ---
names_series = df1[col_name].dropna().astype(str).str.strip().apply(normalize_ar)
name_set = set(names_series)

# --- 5) تأكد من وجود العمود بدون_تشكيل في df_words ---
if 'بدون_تشكيل' not in df_words.columns:
    df_words['بدون_تشكيل'] = df_words['data'].astype(str).apply(normalize_ar)
else:
    df_words['بدون_تشكيل'] = df_words['بدون_تشكيل'].astype(str).apply(normalize_ar)

# --- 6) جهّز أعمدة tag و المحذوف إذا مش موجودة ---
if 'tag' not in df_words.columns:
    df_words['tag'] = pd.NA
if 'المحذوف' not in df_words.columns:
    df_words['المحذوف'] = pd.NA

# --- 7) نعمل مطابقة بين df_words و df1 ---
mask = df_words['بدون_تشكيل'].isin(name_set)

# نصنفها كـ "اسم"
df_words.loc[mask, 'tag'] = 'اسم'

# ننقل الكلمة بدون تشكيل إلى عمود المحذوف
df_words.loc[mask, 'المحذوف'] = df_words.loc[mask, 'بدون_تشكيل']

# نحذفها من عمود بدون_تشكيل
df_words.loc[mask, 'بدون_تشكيل'] = pd.NA

# --- 8) ملخص ---
total = len(df_words)
found = int(mask.sum())
print(f"Total rows in df_words: {total}")
print(f"Matched and classified as 'اسم': {found}")
```

```python
import re

# نسخ الداتا قبل التعديل
df = df_words.copy()

def classify_al_ta3reef(row):
    word = row["data"]
    if pd.isna(word):
        return row
    
    # الشرط الأول: لام + سكون (الْـ)
    if re.match(r"^(?:ٱل|ال)ْ", word):
        row["tag"] = "اسم"
        row["المحذوف"] = word
        row["data"] = None
        return row
    
    # الشرط الثاني: لام بعدها حرف مشدد (الرَّ، الصِّ ...)
    if re.match(r"^(?:ٱل|ال)[^\s]*ّ", word):
        row["tag"] = "اسم"
        row["المحذوف"] = word
        row["data"] = None
        return row
    
    return row

# تطبيق الدالة
df = df.apply(classify_al_ta3reef, axis=1)

# تحديث البيانات
df_words.update(df)

# نشوف الإحصائية
print(df_words["tag"].value_counts())
```

```python
def get_word_pattern(word):
    # Define the patterns (awzan)
    awzan_af3al = [
        "فعل", "فعلَ", "فعلُ", "فعلِ",
        "فَعَّل", "فاعَل", "تَفَعَّل", "تَفاعَل",
        "انفعل", "افتعل", "افعلّ", "استفعل",
        "يفعل", "تفعل", "نفعل", "افعل",
        "مفعل", "مفعول", "مفاعل", "تفعيل"
    ]
    
    awzan_asma = [
        "فاعل", "مفعول", "فعيل", "فعول",
        "فَعَل", "فُعول", "فِعال", "فِعلة",
        "فعلان", "فعلى", "فعيلة", "فعالة",
        "مفعل", "مفعلة", "افتعال", "استفعال",
        "فعائل", "مفاعل", "أفعال", "فعالى"
    ]
    
    awzan = awzan_af3al + awzan_asma
    
    # Function to extract pattern
    def extract_pattern(word, pattern):
        if len(word) != len(pattern):
            return False
        
        # Map to store letter mappings
        letter_map = {}
        pattern_map = {}
        
        for w, p in zip(word, pattern):
            # If we've seen this letter before
            if w in letter_map:
                # Check if it maps to the same pattern letter
                if letter_map[w] != p:
                    return False
            # If we've seen this pattern letter before
            elif p in pattern_map:
                # Check if it maps to the same word letter
                if pattern_map[p] != w:
                    return False
            else:
                # New mapping
                letter_map[w] = p
                pattern_map[p] = w
        
        return True
    
    # Try to match with each pattern
    matches = []
    for pattern in awzan:
        if extract_pattern(word, pattern):
            matches.append(pattern)
    
    return matches if matches else None

# Test the function with some examples
test_words = ["كتب", "يكتب", "مكتوب", "كاتب"]
for word in test_words:
    patterns = get_word_pattern(word)
    print(f"Word: {word}")
    print(f"Matching patterns: {patterns}")
    print()
```

```python
import re

def separate_arabic_letters_and_diacritics(word):
    """Separate Arabic word into letters and diacritics"""
    # Define diacritics (تشكيل)
    diacritics = {
        'فتحة': 'َ',
        'ضمة': 'ُ',
        'كسرة': 'ِ',
        'سكون': 'ْ',
        'شدة': 'ّ',
        'تنوين_فتح': 'ً',
        'تنوين_ضم': 'ٌ',
        'تنوين_كسر': 'ٍ'
    }
    
    # Create regex pattern for all diacritics
    diacritics_pattern = '[' + ''.join(diacritics.values()) + ']'
    
    # Split word into letters and diacritics
    letters = re.sub(diacritics_pattern, '', word)
    diacritics_found = re.findall(diacritics_pattern, word)
    
    return letters, diacritics_found

def create_pattern_template(pattern):
    """Convert a pattern like فاعل into a positional template"""
    letters, diacritics = separate_arabic_letters_and_diacritics(pattern)
    
    # Create mapping of pattern positions
    template = {
        'letters': [],
        'diacritics': diacritics,
        'positions': {}
    }
    
    # Map positions of ف، ع، ل in pattern
    for i, letter in enumerate(letters):
        if letter == 'ف':
            template['positions']['fa'] = i
        elif letter == 'ع':
            template['positions']['ain'] = i
        elif letter == 'ل':
            template['positions']['lam'] = i
        template['letters'].append(letter)
    
    return template

def match_word_to_pattern(word, pattern_template):
    """Try to match a word to a pattern template"""
    word_letters, word_diacritics = separate_arabic_letters_and_diacritics(word)
    
    # Check if length matches
    if len(word_letters) != len(pattern_template['letters']):
        return False, None
    
    # Extract root letters based on pattern positions
    root_letters = {}
    if 'fa' in pattern_template['positions']:
        fa_pos = pattern_template['positions']['fa']
        if fa_pos < len(word_letters):
            root_letters['fa'] = word_letters[fa_pos]
    
    if 'ain' in pattern_template['positions']:
        ain_pos = pattern_template['positions']['ain']
        if ain_pos < len(word_letters):
            root_letters['ain'] = word_letters[ain_pos]
    
    if 'lam' in pattern_template['positions']:
        lam_pos = pattern_template['positions']['lam']
        if lam_pos < len(word_letters):
            root_letters['lam'] = word_letters[lam_pos]
    
    # Check diacritics pattern (if available)
    diacritics_match = True
    if pattern_template['diacritics']:
        if len(word_diacritics) != len(pattern_template['diacritics']):
            diacritics_match = False
        else:
            for w_dia, p_dia in zip(word_diacritics, pattern_template['diacritics']):
                if w_dia != p_dia:
                    diacritics_match = False
                    break
    
    return True, {
        'root_letters': root_letters,
        'word_diacritics': word_diacritics,
        'diacritics_match': diacritics_match
    }

# Test with some patterns
test_patterns = ['فَعَلَ', 'فَاعِل', 'مَفْعُول', 'تَفَعَّلَ']
pattern_templates = {p: create_pattern_template(p) for p in test_patterns}

# Test with a word
test_word = 'كَاتِب'
print(f"Testing word: {test_word}")
print("Pattern matches:")
for pattern, template in pattern_templates.items():
    matches, details = match_word_to_pattern(test_word, template)
    if matches:
        print(f"\nPattern: {pattern}")
        print(f"Root letters: {details['root_letters']}")
        print(f"Word diacritics: {details['word_diacritics']}")
        print(f"Diacritics match: {details['diacritics_match']}")
```

```python
# Define all patterns with their classifications
patterns_data = {
    'verb_patterns': [
        'فَعَلَ', 'فَعَّلَ', 'فَاعَلَ', 'أَفْعَلَ', 'تَفَعَّلَ', 'تَفَاعَلَ',
        'انْفَعَلَ', 'افْتَعَلَ', 'افْعَلَّ', 'اسْتَفْعَلَ',
        'يَفْعَلُ', 'يُفَعِّلُ', 'يُفَاعِلُ', 'يُفْعِلُ'
    ],
    'noun_patterns': [
        'فَاعِل', 'مَفْعُول', 'فَعِيل', 'فَعُول',
        'فَعَل', 'فُعُول', 'فِعَال', 'فِعْلَة',
        'فَعْلَان', 'فَعْلَى', 'فَعِيلَة', 'فَعَالَة',
        'مَفْعَل', 'مَفْعَلَة', 'افْتِعَال', 'اسْتِفْعَال'
    ]
}

# Create templates for all patterns
pattern_templates = {}
for category, patterns in patterns_data.items():
    for pattern in patterns:
        pattern_templates[pattern] = {
            'template': create_pattern_template(pattern),
            'category': category
        }

def analyze_word_patterns(word):
    """Analyze a word against all known patterns"""
    results = []
    
    for pattern, data in pattern_templates.items():
        matches, details = match_word_to_pattern(word, data['template'])
        if matches and details['diacritics_match']:
            results.append({
                'pattern': pattern,
                'category': data['category'],
                'root_letters': details['root_letters'],
                'diacritics': details['word_diacritics']
            })
    
    return results

# Function to update DataFrame with pattern analysis
def analyze_and_update_patterns(df):
    df = df.copy()
    
    def process_word(row):
        if pd.isna(row['data']):
            return row
        
        # Analyze patterns
        patterns = analyze_word_patterns(row['data'])
        
        if patterns:
            # Store pattern information
            row['patterns_found'] = patterns
            
            # Determine if it's a verb or noun based on patterns
            is_verb = any(p['category'] == 'verb_patterns' for p in patterns)
            is_noun = any(p['category'] == 'noun_patterns' for p in patterns)
            
            # Update classification if not already classified
            if pd.isna(row['tag']):
                if is_verb and not is_noun:
                    row['tag'] = 'فعل'
                    row['المحذوف'] = row['data']
                    row['data'] = None
                elif is_noun and not is_verb:
                    row['tag'] = 'اسم'
                    row['المحذوف'] = row['data']
                    row['data'] = None
                    
            # Store root letters if found
            if patterns[0]['root_letters']:
                row['root_letters'] = patterns[0]['root_letters']
        
        return row
    
    # Apply the analysis to each row
    df = df.apply(process_word, axis=1)
    
    return df

# Process a sample of words
sample_df = df_words[df_words['data'].notna()].head(20).copy()
analyzed_df = analyze_and_update_patterns(sample_df)

# Display results
print("\nPattern Analysis Results:")
for idx, row in analyzed_df.iterrows():
    if 'patterns_found' in row and not pd.isna(row['patterns_found']):
        print(f"\nWord: {row['المحذوف'] or row['data']}")
        print("Matching patterns:")
        for p in row['patterns_found']:
            print(f"- Pattern: {p['pattern']} ({p['category']})")
            print(f"  Root letters: {p['root_letters']}")
            print(f"  Diacritics: {p['diacritics']}")
```

```python
# Add position information to our patterns
def enhance_pattern_info(pattern):
    """Add more detailed information about pattern structure"""
    template = create_pattern_template(pattern)
    
    # Get prefix letters (before fa)
    prefix_letters = []
    for i, letter in enumerate(template['letters']):
        if letter == 'ف':
            break
        prefix_letters.append(letter)
    
    # Get suffix letters (after lam)
    suffix_letters = []
    lam_pos = None
    for i, letter in enumerate(template['letters']):
        if letter == 'ل':
            lam_pos = i
    if lam_pos is not None:
        suffix_letters = template['letters'][lam_pos + 1:]
    
    return {
        'template': template,
        'prefix': ''.join(prefix_letters),
        'suffix': ''.join(suffix_letters),
        'diacritics_pattern': template['diacritics']
    }

# Enhance our pattern templates
enhanced_patterns = {}
for category, patterns in patterns_data.items():
    for pattern in patterns:
        enhanced_patterns[pattern] = {
            **enhance_pattern_info(pattern),
            'category': category
        }

def analyze_word_enhanced(word):
    """Enhanced word analysis with more detailed pattern matching"""
    letters, diacritics = separate_arabic_letters_and_diacritics(word)
    results = []
    
    for pattern, data in enhanced_patterns.items():
        template = data['template']
        if len(letters) == len(template['letters']):
            matches, details = match_word_to_pattern(word, template)
            if matches:
                match_info = {
                    'pattern': pattern,
                    'category': data['category'],
                    'root_letters': details['root_letters'],
                    'diacritics': details['word_diacritics'],
                    'prefix': data['prefix'],
                    'suffix': data['suffix'],
                    'diacritics_match': details['diacritics_match']
                }
                results.append(match_info)
    
    return results

# Test with a sample word
test_words = ['كَتَبَ', 'يَكْتُبُ', 'مَكْتُوب', 'كَاتِب', 'كِتَاب']
print("Enhanced Pattern Analysis:")
for word in test_words:
    print(f"\nAnalyzing word: {word}")
    results = analyze_word_enhanced(word)
    for r in results:
        print(f"\nPattern: {r['pattern']} ({r['category']})")
        print(f"Root letters: {r['root_letters']}")
        print(f"Diacritics: {r['diacritics']}")
        print(f"Prefix: {r['prefix']}")
        print(f"Suffix: {r['suffix']}")
        print(f"Diacritics match: {r['diacritics_match']}")
```

```python
def process_remaining_words(df):
    """Process remaining unclassified words in the DataFrame"""
    df = df.copy()
    
    # Add columns for pattern information if they don't exist
    if 'pattern_info' not in df.columns:
        df['pattern_info'] = None
    if 'root_letters' not in df.columns:
        df['root_letters'] = None
    
    def process_row(row):
        if pd.isna(row['tag']) and pd.notna(row['data']):
            # Analyze the word
            patterns = analyze_word_enhanced(row['data'])
            
            if patterns:
                # Store all pattern matches
                row['pattern_info'] = patterns
                
                # Check if we can classify the word
                verb_patterns = [p for p in patterns if p['category'] == 'verb_patterns' and p['diacritics_match']]
                noun_patterns = [p for p in patterns if p['category'] == 'noun_patterns' and p['diacritics_match']]
                
                if verb_patterns and not noun_patterns:
                    row['tag'] = 'فعل'
                    row['المحذوف'] = row['data']
                    row['data'] = None
                    # Store root letters from the first matching pattern
                    row['root_letters'] = verb_patterns[0]['root_letters']
                
                elif noun_patterns and not verb_patterns:
                    row['tag'] = 'اسم'
                    row['المحذوف'] = row['data']
                    row['data'] = None
                    # Store root letters from the first matching pattern
                    row['root_letters'] = noun_patterns[0]['root_letters']
        
        return row
    
    # Process all rows
    df = df.apply(process_row, axis=1)
    
    return df

# Process remaining words
df_words = process_remaining_words(df_words)

# Show statistics
print("Updated classification statistics:")
print(df_words['tag'].value_counts())
print("\nWords with identified root letters:", df_words['root_letters'].notna().sum())

# Show some examples of classified words with their patterns
print("\nExamples of classified words with pattern information:")
sample = df_words[df_words['pattern_info'].notna()].head(10)
for idx, row in sample.iterrows():
    print(f"\nWord: {row['المحذوف'] or row['data']}")
    print(f"Tag: {row['tag']}")
    if isinstance(row['pattern_info'], list):
        for pattern in row['pattern_info']:
            print(f"- Pattern: {pattern['pattern']} ({pattern['category']})")
            print(f"  Root letters: {pattern['root_letters']}")
            print(f"  Diacritics match: {pattern['diacritics_match']}")
```

```python
def analyze_word_with_prefixes(word):
    """Analyze a word considering common prefixes and special cases"""
    letters, diacritics = separate_arabic_letters_and_diacritics(word)
    
    # Special cases patterns
    special_cases = {
        'ال': r'^ال',
        'بِ': r'^ب\u0650',  # ب with kasra
        'لِ': r'^ل\u0650',  # ل with kasra
        'وَ': r'^و\u064E',  # و with fatha
        'فَ': r'^ف\u064E',  # ف with fatha
    }
    
    # Check for prefixes
    prefixes_found = []
    remaining_word = word
    
    # First check for compound prefixes (e.g., بال، وال، فال)
    compound_prefixes = {
        'بِال': r'^ب\u0650ال',
        'وَال': r'^و\u064Eال',
        'فَال': r'^ف\u064Eال',
        'لِل': r'^لِل',
    }
    
    for prefix, pattern in compound_prefixes.items():
        if re.match(pattern, word):
            prefixes_found.append(prefix)
            remaining_word = re.sub(pattern, '', word)
            break
    
    # If no compound prefix found, check for single prefixes
    if not prefixes_found:
        for prefix, pattern in special_cases.items():
            if re.match(pattern, remaining_word):
                prefixes_found.append(prefix)
                remaining_word = re.sub(pattern, '', remaining_word)
    
    # Analyze the remaining word
    base_analysis = analyze_word_enhanced(remaining_word) if remaining_word else []
    
    return {
        'original_word': word,
        'prefixes': prefixes_found,
        'remaining_word': remaining_word,
        'base_patterns': base_analysis,
        'letters': letters,
        'diacritics': diacritics
    }

# Test with some sample words
test_words = ['بِسْمِ', 'اللَّهِ', 'الرَّحْمَٰنِ', 'الرَّحِيمِ', 'الْحَمْدُ', 'لِلَّهِ']
print("Enhanced Analysis with Prefixes:")
for word in test_words:
    analysis = analyze_word_with_prefixes(word)
    print(f"\nWord: {word}")
    print(f"Prefixes found: {analysis['prefixes']}")
    print(f"Remaining word: {analysis['remaining_word']}")
    print(f"Letters: {analysis['letters']}")
    print(f"Diacritics: {analysis['diacritics']}")
    if analysis['base_patterns']:
        print("Matching patterns:")
        for pattern in analysis['base_patterns']:
            print(f"- {pattern['pattern']} ({pattern['category']})")
            print(f"  Root letters: {pattern['root_letters']}")
    else:
        print("No base patterns matched")
```

# Arabic Word Pattern Analysis Algorithm

Our algorithm analyzes Arabic words by comparing them with traditional Arabic word patterns (أوزان). Here's how it works:

## 1. Pattern Matching Process
1. **Prefix Detection**:
   - First checks for common prefixes like ال، بال، وال، فال
   - Removes prefix to analyze the base word

2. **Letter Mapping**:
   - Maps the word's letters to the pattern's letters (ف ع ل)
   - For example: كَتَبَ maps to فَعَلَ
   - Maintains consistent mapping (same letter always maps to same pattern position)

3. **Diacritics Analysis**:
   - Separates letters from diacritics (حركات)
   - Compares diacritic patterns separately

## 2. Classification Rules
1. **Special Cases**:
   - Some words (like الله، رب) are directly classified
   - Words with ال prefix are classified as nouns (اسم)

2. **Pattern-Based Classification**:
   - Verb Patterns (أوزان الأفعال): فعل، فعّل، فاعل، etc.
   - Noun Patterns (أوزان الأسماء): فاعل، مفعول، فعيل، etc.

3. **Root Extraction**:
   - Identifies root letters based on their position in matched patterns
   - Maps back to ف ع ل positions

## 3. Results Storage
- Tags words as فعل (verb) or اسم (noun)
- Stores the original word in المحذوف column
- Records root letters when found
- Maintains pattern information for reference

```python
def get_letter_positions(word):
    """Get positions of each letter in the word."""
    positions = {}
    current_pos = 0
    
    # Remove diacritics for position mapping
    clean_word = re.sub(r'[ًٌٍَُِّْ]', '', word)
    
    for i, letter in enumerate(clean_word):
        positions[letter] = positions.get(letter, []) + [i]
    
    return positions

def pattern_match_score(word_positions, pattern_positions):
    """Calculate how well a pattern matches a word based on letter positions."""
    score = 0
    total_positions = 0
    
    # Check if key positions match (like م in position 1 for مفعول)
    for letter, positions in pattern_positions.items():
        if letter in word_positions:
            for pos in positions:
                if pos in word_positions[letter]:
                    score += 1
                total_positions += 1
    
    return score / total_positions if total_positions > 0 else 0

def find_best_pattern_match(word):
    """Find the best matching pattern for a word."""
    # Define common patterns with their expected letter positions
    patterns = {
        'مفعول': {
            'م': [0],  # م at start
            'ف': [1],  # First root letter
            'ع': [2],  # Second root letter
            'و': [3],  # و characteristic of مفعول
            'ل': [4]   # Third root letter
        },
        'فَعَلَ': {
            'ف': [0],
            'ع': [1],
            'ل': [2]
        },
        # Add more patterns here
    }
    
    # Get positions of letters in the input word
    word_positions = get_letter_positions(word)
    
    # Calculate match score for each pattern
    scores = {}
    for pattern_name, pattern_pos in patterns.items():
        score = pattern_match_score(word_positions, pattern_pos)
        scores[pattern_name] = score
    
    # Get the best matching pattern
    best_pattern = max(scores.items(), key=lambda x: x[1])
    
    return best_pattern[0] if best_pattern[1] > 0.5 else None  # Return pattern only if score > 50%

# Test with مكتوب
test_word = 'مَكْتُوب'
best_pattern = find_best_pattern_match(test_word)
print(f"Word: {test_word}")
print(f"Best matching pattern: {best_pattern}")
```
